```python
import MEArec as mr
import numpy as np
import scipy.optimize
import os
import sys
import re
import ast
import matplotlib.pyplot as plt
import pandas as pd
import pickle
import time
from collections import defaultdict
from joblib import Parallel, delayed

import spikeinterface as si
import spikeinterface.core as sc
import spikeinterface.extractors as se
import spikeinterface.sorters as ss
import spikeinterface.preprocessing as spre
import spikeinterface.postprocessing as spost
import spikeinterface.widgets as sw
import spikeinterface.comparison as scomp
import spikeinterface.curation as scu

sys.path.append('src')
from src.util_eval import *
from src.util_loc import *
```

```python
# Overall parameters
mearec_seed = 42
dead_indices_seeds = [42, 43, 44, 45, 46]
run_id = f'20260210_{mearec_seed}'
sorter_name = 'mountainsort4'
days = ['D0', 'D1', 'D2', 'D3', 'D4', 'D5', 'D6', 'D7', 'D8', 'D9',
        'D10','D11','D12','D13','D14','D15','D16','D17','D18','D19',
        'D20','D21','D22','D23','D24']
methods = ['center_of_mass', 'monopolar_triangulation', 'grid_convolution']
stable_days = 1
dead_electrodes_per_day = 15
gt_toggle = 0 # if 0, use ground truth data as experimental data
correct_radius = 30 # microns

# MEArec parameters
mearec_probe = 'Neuropixels-384'
num_channels = 384
recgen_duration = 60
mearec_noise_level = 10
output_mearec = f'output/{run_id}'
```

```python
# Create the template and recording generator for multiple days
cell_folder = mr.get_default_cell_models_folder()
temp_params = mr.get_default_templates_params()
temp_params['n'] = 25
temp_params['probe'] = mearec_probe

for dead_indices_seed_i, dead_indices_seed in enumerate(dead_indices_seeds):    
    output_mearec_seed = f'{output_mearec}/seed_{dead_indices_seed}'
    
    if os.path.exists(f'{output_mearec_seed}/templates.h5'):
        tempgen = mr.load_templates(f'{output_mearec_seed}/templates.h5')
        recgen = mr.load_recordings(f'{output_mearec_seed}/D0/recording.h5')
    
    else:
        temp_params['seed'] = mearec_seed
        tempgen = mr.gen_templates(cell_models_folder=cell_folder, params=temp_params, templates_tmp_folder=None, intraonly=False, 
                                   parallel=True, recompile=False, n_jobs=None, delete_tmp=True, verbose=False)
        mr.save_template_generator(tempgen, f'{output_mearec_seed}/templates.h5')
            
        for day_i, day in enumerate(days):
            rec_params = mr.get_default_recordings_params()
            rec_params['spiketrains']['n_exc'] = 35 # Number of excitatory neurons
            rec_params['spiketrains']['n_inh'] = 15 # Number of inhibitory neurons

            rec_params['spiketrains']['duration'] = recgen_duration # Duration of recording in seconds
            rec_params['recordings']['fs'] = 30000 # Sampling frequency
            rec_params['recordings']['noise_level'] = mearec_noise_level
            rec_params['seeds']['templates'] = mearec_seed

            recgen = mr.gen_recordings(params=rec_params, tempgen=tempgen)
            mr.save_recording_generator(recgen, f'{output_mearec_seed}/{day}/recording.h5')
```

```python
# Function to process each seed
def process_seed(dead_indices_seed_i, dead_indices_seed):
    output_mearec_seed = f'{output_mearec}/seed_{dead_indices_seed}'
    
    # Dead indices
    if os.path.exists(f'{output_mearec_seed}/dead_indices.pkl'):
        print('Loading dead indices')
        with open(f'{output_mearec_seed}/dead_indices.pkl', 'rb') as f:
            dead_indices = pickle.load(f)
    else:
        print('Creating dead indices')
        np.random.seed(dead_indices_seed)
        
        dead_indices = []
        dead_indices_temp = np.random.choice(range(num_channels), size=dead_electrodes_per_day*(len(days)-stable_days), replace=False)
        for day_i, day in enumerate(days):
            if day_i < stable_days:
                dead_indices.append([])
            else:
                dead_indices.append(dead_indices_temp[:dead_electrodes_per_day*(day_i-1)])
        
        with open(f'{output_mearec_seed}/dead_indices.pkl', "wb") as file:
            pickle.dump(dead_indices, file)

    # Waveform extractors
    if os.path.exists(f'{output_mearec_seed}/D0/waveforms_gt'):
        print('Loading waveform extractors')
        
        wes_gt = []
        for day_i, day in enumerate(days):
            wes_gt.append(si.load_waveforms(f'{output_mearec_seed}/{day}/waveforms_gt', with_recording=False))
        
    else:    
        # Recording
        recording_mearec = se.MEArecRecordingExtractor(f'{output_mearec_seed}/D0/recording.h5')
        probe = recording_mearec.get_probe()

        recordings = []
        for day_i, day in enumerate(days):
            
            if os.path.exists(f'{output_mearec_seed}/{day}/recording'):
                recordings.append(sc.load_extractor(f'{output_mearec_seed}/{day}/recording').set_probe(probe, in_place=True))
            else:
                recording_mearec = se.MEArecRecordingExtractor(f'{output_mearec_seed}/{day}/recording.h5')
                probe = recording_mearec.get_probe()
                get_recording_noise(recording_mearec, dead_indices[day_i], f'{output_mearec_seed}/{day}/recording')
                recordings.append(sc.load_extractor(f'{output_mearec_seed}/{day}/recording').set_probe(probe, in_place=True))
            
        # Sorting data
        sortings_gt = []
        for day_i, day in enumerate(days):
            sorting_mearec = se.MEArecSortingExtractor(f'{output_mearec_seed}/{day}/recording.h5')
            
            # Extract all the spike data
            unit_spike_dict = {}
            for unit_id in sorting_mearec.get_unit_ids():
                spike_times = sorting_mearec.get_unit_spike_train(unit_id)
                unit_spike_dict[unit_id] = spike_times

            # Create NumpySorting (no file dependencies)
            numpy_sorting = sc.NumpySorting.from_unit_dict(
                unit_spike_dict,
                sampling_frequency=sorting_mearec.get_sampling_frequency()
            )

            # Copy any properties if needed
            for prop_key in sorting_mearec.get_property_keys():
                numpy_sorting.set_property(prop_key, sorting_mearec.get_property(prop_key))

            sortings_gt.append(numpy_sorting)

        # Waveform extractor data
        wes_gt = []
        for day_i, day in enumerate(days):
            wes_gt.append(si.extract_waveforms(recordings[day_i], sortings_gt[day_i], folder=f'{output_mearec_seed}/{day}/waveforms_gt', ms_before=1, ms_after=2, load_if_exists=True, use_relative_path=True))
            wes_gt[-1].run_extract_waveforms()

    # Calculate template locations
    if os.path.exists(f'{output_mearec_seed}/loc_est_units'):
        print('Template localization complete')
            
    else:
        print('Calculating template localization estimates.')
        os.makedirs(f'{output_mearec_seed}/loc_est_units')
        
        # Base parameters        
        loc_est_units = {} # First entry is method. Then the entry is a list of locations across days.
        time_units = {}

        for method in methods:
            start_time = time.time()
            loc_est_units[method] = get_unit_loc_est(method, wes_gt)
            time_units[method] = time.time() - start_time # time to go through all days
        
        pickle.dump(loc_est_units, open(f'{output_mearec_seed}/loc_est_units/loc_est_units.pkl', 'wb'))
        pickle.dump(time_units, open(f'{output_mearec_seed}/loc_est_units/time_units.pkl', 'wb'))        

        # Grid search parameters
        mt_param_1 = [37.5, 75, 150] # radius_um
        mt_param_2 = [500, 1000, 2000] # max_distance_um

        gc_param_1 = [20, 40, 80] # radius_um
        gc_param_2 = [2.5, 5, 10] # percentile        
        
        loc_est_grid_units = {} # First and second entries are parameters. Then method. Then list of locations across days.
        time_grid_units = {}

        for param_1 in range(3):
            loc_est_grid_units[param_1] = {}
            time_grid_units[param_1] = {}
            
            for param_2 in range(3):
                loc_est_grid_units[param_1][param_2] = {}
                time_grid_units[param_1][param_2] = {}
                
                for method in methods:
                    start_time = time.time()
                    if method == 'monopolar_triangulation':
                        loc_est_grid_units[param_1][param_2][method] = get_unit_loc_grid(method, wes_gt, grid_params={'radius_um': mt_param_1[param_1], 'max_distance_um': mt_param_2[param_2]}, days=days)
                    elif method == 'grid_convolution':
                        loc_est_grid_units[param_1][param_2][method] = get_unit_loc_grid(method, wes_gt, grid_params={'radius_um': gc_param_1[param_1], 'percentile': gc_param_2[param_2]}, days=days)
                    else: # COM; don't need to calculate to save time
                        loc_est_grid_units[param_1][param_2][method] = None
                        
                    time_grid_units[param_1][param_2][method] = time.time() - start_time # time to go through all days

        pickle.dump(loc_est_grid_units, open(f'{output_mearec_seed}/loc_est_units/loc_est_grid_units.pkl', 'wb'))
        pickle.dump(time_grid_units, open(f'{output_mearec_seed}/loc_est_units/time_grid_units.pkl', 'wb'))

    # Calculate spike location estimates
    if os.path.exists(f'{output_mearec_seed}/loc_est_spikes'):
        print('Spike localization complete')

    else:
        print('Calculating spike localization estimates')
        os.makedirs(f'{output_mearec_seed}/loc_est_spikes')

        # Base parameters        
        loc_est_spikes = {} # First entry is method, second is day.
        time_spikes = {}

        for method in methods:
            loc_est_spikes[method] = {}
            time_spikes[method] = {}
            for day_i, day in enumerate(days):
                
                start_time = time.time()
                if method == 'monopolar_triangulation':
                    unit_loc_est = spost.compute_spike_locations(wes_gt[day_i], method=method, outputs='by_unit', method_kwargs={'optimizer': 'least_square'})
                else:
                    unit_loc_est = spost.compute_spike_locations(wes_gt[day_i], method=method, outputs='by_unit')
                
                time_spikes[method][day] = time.time() - start_time # time to go through one day                
                loc_est_spikes[method][day] = unit_loc_est
        
        pickle.dump(loc_est_spikes, open(f'{output_mearec_seed}/loc_est_spikes/loc_est_spikes.pkl', 'wb'))
        pickle.dump(time_spikes, open(f'{output_mearec_seed}/loc_est_spikes/time_spikes.pkl', 'wb'))

        # Grid search parameters
        mt_param_1 = [37.5, 75, 150] # radius_um
        mt_param_2 = [75, 150, 300] # max_distance_um

        gc_param_1 = [20, 40, 80] # radius_um
        gc_param_2 = [2.5, 5, 10] # percentile        
        
        loc_est_grid_spikes = {} # First entry is method, second is day.
        time_grid_spikes = {}
        
        for param_1 in range(3):
            loc_est_grid_spikes[param_1] = {}
            time_grid_spikes[param_1] = {}
            
            for param_2 in range(3):
                loc_est_grid_spikes[param_1][param_2] = {}
                time_grid_spikes[param_1][param_2] = {}

                for method in methods:
                    loc_est_grid_spikes[param_1][param_2][method] = {}
                    time_grid_spikes[param_1][param_2][method] = {}
                    
                    for day_i, day in enumerate(days):
                        if day in ['D0', 'D8', 'D16', 'D24']:
                            
                            start_time = time.time()
                            if method == 'monopolar_triangulation':
                                spike_loc_est = spost.compute_spike_locations(wes_gt[day_i], method=method, outputs='by_unit', method_kwargs={'optimizer': 'least_square', 'radius_um': mt_param_1[param_1], 'max_distance_um': mt_param_2[param_2]})
                            elif method == 'grid_convolution':
                                spike_loc_est = spost.compute_spike_locations(wes_gt[day_i], method=method, outputs='by_unit', method_kwargs={'radius_um': gc_param_1[param_1], 'percentile': gc_param_2[param_2]})
                            else:
                                spike_loc_est = None
                            
                            time_grid_spikes[param_1][param_2][method][day] = time.time() - start_time # time to go through one day
                            loc_est_grid_spikes[param_1][param_2][method][day] = spike_loc_est

        pickle.dump(loc_est_grid_spikes, open(f'{output_mearec_seed}/loc_est_spikes/loc_est_grid_spikes.pkl', 'wb'))
        pickle.dump(time_grid_spikes, open(f'{output_mearec_seed}/loc_est_spikes/time_grid_spikes.pkl', 'wb'))

    # Delete the day subfolders except first and all data inside
    for day_i, day in enumerate(days[1:]):
        os.system(f'rm -r {output_mearec_seed}/{day}/recording')
        os.system(f'rm -r {output_mearec_seed}/{day}/recording.h5')

# Processing seeds
for dead_indices_seed_i, dead_indices_seed in enumerate(dead_indices_seeds):
    process_seed(dead_indices_seed_i, dead_indices_seed)
```
